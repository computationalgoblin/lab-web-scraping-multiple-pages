```python
from bs4 import BeautifulSoup
import httpx
from selectolax.parser import HTMLParser
import pandas as pd 
import numpy as np

```

```python
url = "https://www.billboard.com/charts/hot-100/"

resp = httpx.get(url)
html = HTMLParser(resp.text)
html
```

```python
list = html.css("li.o-chart-results-list__item")
print(list)
```

```python
titles = []

for title in list:
    if title.css_first(".c-title") != None:
        titles.append(title.css_first(".c-title").text().strip())
titles
```

```python
artists = []

for artist in list:
    if artist.css_first("span.c-label") != None:
        artists.append(artist.css_first("span.c-label").text().strip())
artists
```

```python
artists = [artist for artist in artists if not artist.isdigit()]
artists = [artist for artist in artists if artist not in ["-", "NEW",'RE-\nENTRY']]
artists
```

```python
data = pd.DataFrame(zip(titles, artists), columns=["song_name", "artist_name"])
billboard = data.applymap(lambda x: x.lower().replace(" ","_"))
billboard
```

```python
url_1 = "https://en.wikipedia.org/wiki/List_of_Billboard_Alternative_Airplay_number_ones_of_the_2020s#2024"
```

```python
resp = httpx.get(url_1)
html = HTMLParser(resp.text)
html
```

```python
table = html.css("table.plainrowheaders tbody tr")
rows = []
print(table)
```

```python
for row in table:
    cells = row.css("th") + row.css("td a")
    row_data = [cell.text().strip() for cell in cells]
    rows.append(row_data)

rows
```

```python
wiki_data = pd.DataFrame(data = rows)
wiki_data = wiki_data.drop(columns= [2,3,4])
wiki_data.columns = ["song_title", "artist"]
wiki_data = wiki_data.dropna()
wiki_data

pd.set_option('display.max_rows', None)
```

```python
wiki_data = wiki_data.drop_duplicates()
wiki_data = wiki_data.drop(index=0)
```

```python
pd.set_option('display.max_rows', 20)
```

```python
wiki_data.rename(columns={'song_title': 'song_name', 'artist': 'artist_name'}, inplace=True)

display(wiki_data)
display(billboard)
```

```python
final_data = pd.concat([wiki_data, billboard], axis=0)
final_data.reset_index(drop=True, inplace=True)
final_data = final_data.applymap(lambda x: x.lower().replace(" ", "_"))
final_data
```
